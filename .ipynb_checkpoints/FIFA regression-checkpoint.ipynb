## Overall vs wage

```python
x = df["Overall"]
y = df["Wage"]
player_wage = []
for i in range(len(y)):
    if y[i] != '':
        tmp = y[i]
        tmp = tmp[1:]
        unit_of_m = tmp[-1]
        if unit_of_m in "MK":
            tmp = tmp[:-1]
        if tmp != '':
            money = float(tmp)
        if unit_of_m == 'M':
            money = money * 1000

        player_wage.append(money)
    else:
        player_wage.append(0)
plt.figure(figsize=(9,14))
plt.title("Overall vs wage scatter plt")
plt.scatter(x,player_wage,s=8)
```

## Regression overall vs wage

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x = np.array(x)
x = x.reshape(len(x),1)
x_train,x_test,y_train,y_test = train_test_split(x,player_wage, test_size=0.2, random_state = 705023)
regr = LinearRegression()
regr.fit(x_train,y_train)

Y_pred = regr.predict(x_test)
plt.figure(figsize=(9,14))
plt.scatter(x_test, y_test,s=8)
plt.plot(x_test, Y_pred, 'r')
print("regression score:"+str(regr.score(x_test, y_test)))
```
