# Face recognition and segmenting in videos
In this kernel, I will be demonstrating how to do facial clustering in a given video/ image. For this case, I will be using an open source project called *face_recognition* by **ageitgey**. The other packages for computer vision is **opencv**.

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
import face_recognition as fr
import dlib
import time
import pandas as pd
dlib.DLIB_USE_CUDA
```

```python
cap = cv2.VideoCapture('y2mate.com - bts_idol_kpop_dance_in_public_VxObfw3Ruew_1080p.mp4')


# Check if camera opened successfully
if (cap.isOpened()== False): 
  print("Error opening video stream or file")
 
# Read until video is completed
cnt = 1
imgs = []
while(cap.isOpened()):
    # Capture frame-by-frame
    ret, frame = cap.read()
    if cnt % 10 == 5:
        
        img = frame
        img = img[...,::-1]
        imgs.append(img)
    cnt += 1
    if cnt % 500 == 0:
        print(cnt)
    if ret != True:
        break
 
# When everything done, release the video capture object
cap.release()
print("total frames captured:",len(imgs),sep='')
```

```python
plt.imshow(imgs[326])
```

The above is a image during the video, now we can bound out faces in the picture.

```python
loc = fr.face_locations(imgs[326],model = "cnn",number_of_times_to_upsample=1)
print(len(loc))
for i in range(0,len(loc)):
    print(loc[i][0])
    plt.axis()
    plt.imshow(imgs[326][loc[i][0]:loc[i][2]+1, loc[i][3]:loc[i][1],:])
    plt.show()
```

The above result shows that some of them just doesn't really seems to be working good. For example, I can barely tell it is a face from the picture. 

## Improvements
Well, some improvements can definitely be done simply by selecting some more simpler videos. By **simple**, I mean the less people involved in the video the better.

```python
cap = cv2.VideoCapture('cnbcs_samsung.mp4')


# Check if camera opened successfully
if (cap.isOpened()== False): 
  print("Error opening video stream or file")
 
# Read until video is completed
cnt = 1
imgs = []
while(cap.isOpened()):
    # Capture frame-by-frame
    ret, frame = cap.read()
    if cnt % 10 == 5:
        
        img = frame
        img = img[...,::-1]
        imgs.append(img)
    cnt += 1
    if cnt % 500 == 0:
        print(cnt)
    if ret != True:
        break
 
# When everything done, release the video capture object
cap.release()
print("total frames captured:",len(imgs),sep='')
```

```python
plt.imshow(imgs[100])
plt.show()
loc = fr.face_locations(imgs[100],model = "cnn",number_of_times_to_upsample=1)
print(len(loc))
for i in range(0,len(loc)):
    print(loc[i][0])
    plt.axis()
    plt.imshow(imgs[100][loc[i][0]:loc[i][2]+1, loc[i][3]:loc[i][1],:])
    plt.show()
```

```python

encodings = []
start = time.time()
for i in range(len(imgs)):
    locations = []
    locations.append(fr.face_locations(imgs[i],model = "cnn"))
    for j in locations:
        encodings.append(fr.face_encodings(imgs[i],j))
print("time spent", time.time() - start,sep='')
```

For each image inside what we have collected, we locate all the faces deteced and pass those into a encoding network (A CNN with its bottom Fully Connected layer removed).

```python
print(encodings[100])
```

The two matrix printed above is a 128 dimension vector. This vector can be regarded as the output of passing the facial feature into **ResNet**. This specific matrix is what we acquire before entering to the Fully Connected Layer. Now, I will merge all the encodings inside *encodings* into an array called *enc*.

```python
enc = []
for i in range(len(encodings)):
    for j in range(len(encodings[i])):
        enc.append(encodings[i][j])
print(len(np.concatenate(enc)))
```

The above block helps us encode each face. Now, if we want to do further more classification problems, say PCA(covered down below), k-means, can all be implemented down below.

## PCA
In the following sections, I would like to do some basic classification, without caring each vector being correspondant to which individual. I will only be focusing on how many **potential** identities are inside here.

```python
# print(encodings)


from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(enc)
PCA_df = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
PCA_df.head()
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
colors = 'b'

for i in range(len(enc)):
    ax.scatter(PCA_df.loc[i, 'principal component 1']
               , PCA_df.loc[i, 'principal component 2']
               , c = colors
               , s = 7)
ax.grid()
```

Now that we have completed our PCA analysis, it seems that there are roughly 3 people invovled in the video

## Conclusion, Improvements...

As far as my understanding, there are some shortcomings using the github open-source project.    
* Un able to run multiple images in parallel.
    * When the image becomes a relatively large dataset, I will need to run pictures in batches. Otherwise, I will spend a lot of time waiting and without maxing out what a decent P100 GPU can do  
* Some faces are actually not detected in pictures.
    * Naively, I should have scale down the image by a certain amount, but I didn't. It is creating a immense overhead memory during runtime.
* Also, the implementation efficiency of memory in dlib or the github project, seems to be poorly configured.

### PCA's rough idea vs K-means clustering.
In the field of unsupurvised learning, it is sometimes harder than we think to provide us a feasible solution that is somehow optimal. However, through PCA, in this case, help us visualize a high dimension problem into a relatively comprehensible one.
